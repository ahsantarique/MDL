```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate, optimize
```

```python
df = pd.read_csv('data/georgia_reported.csv')
df.head()
```

```python
ycases = df['cases'].to_numpy()
ydeaths = df['deaths'].to_numpy()
ydata = [(ycases[i], ydeaths[i]) for i in range(len(ycases))]
# ydata = np.concatenate(:)
ydata = np.array(ydata, dtype = float)


xdata = list(range(1, len(ycases)+1))
# xdata = np.array(xdata, dtype=float)
```

```python
def sird_model(y, x, beta, gamma, mu):
    dS = -beta * y[0] * y[1] / N 
    dR = gamma * y[1]
    dM = mu * y[1]
    dI = -(dS + dR + dM)
    return dS, dI, dR, dM
```

```python
def fit_odeint(x, beta, gamma, mu, S0, I0, R0, M0):
    return integrate.odeint(func = sird_model, y0 = (S0, I0, R0, M0), t = x, args=(beta, gamma, mu))[:,1]
    
def fit_odeint_m(x, beta, gamma, mu, S0, I0, R0, M0):
    return integrate.odeint(func = sird_model, y0 = (S0, I0, R0, M0), t = x, args=(beta, gamma, mu))[:,3]


N = 3989167  ## it should be the population... not sure...

popt1, pcov1 = optimize.curve_fit(f = fit_odeint, xdata = xdata, ydata = ycases, p0 = (0.1, 0.1, 0.1, ycases[-1], 1, 0, ydeaths[0]))
popt2, pcov2 = optimize.curve_fit(f = fit_odeint_m, xdata = xdata, ydata = ydeaths, p0 = (0.1, 0.1, 0.1, ycases[-1], 1, 0, ydeaths[0]))

fitted = fit_odeint(xdata, *popt1)



plt.plot(xdata, ycases, 'o')
plt.plot(xdata, fitted)
plt.show()






popt2, pcov2 = optimize.curve_fit(f = fit_odeint, xdata = xdata, ydata = ydeaths, p0 = (0.1, 0.1, 0.1, N, 1, 0, ydeaths[0]))

fitted = fit_odeint(xdata, *popt2)

print(popt2)

plt.plot(xdata, ydeaths, 'o')
plt.plot(xdata, fitted)
plt.show()


print('beta ', popt1[0])
print('gamma ', popt1[1])
print('mu ', popt2[2])
print('S0 ', popt1[3])
print('I0 ', popt1[4])
print('R0', popt1[5])
print('D0', popt2[6])
```
